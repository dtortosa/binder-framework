```python
##import required packages
#for plotting
import matplotlib as mpl
import matplotlib.pyplot as plt
#for data procesing and numerical operations
import pandas as pd
import numpy as np
#for date processing
from datetime import datetime
```

```python
##read data
energy_data = pd.read_csv("data/Value_of_Energy_Cost_Savings_Program_Savings_for_Businesses_-_FY2020.csv")
energy_data
```

```python
##read results
energy_results = pd.read_csv("results.csv", index_col=0) 
    #original row names at first column
energy_results
```

```python
##scatter job vs. savings standardad scale

#open the plot
fig, ax = plt.subplots(figsize=(4, 4))

#select the data and presentation
plt.scatter(x="Average Savings", y="Total Jobs", data=energy_results, s=40, alpha=1)
    #s is for point size; alpha is for color contrast; c is for color

#set the axis labels
ax.set_xlabel("Average Savings")
ax.set_ylabel("Total Jobs")

#set title
ax.set_title("Jobs versus Savings")
```

```python
##scatter job vs. savings log scale

#open the plot
fig, ax = plt.subplots(figsize=(4, 4))

#define the variables to plot
x = np.log(energy_results["Average Savings"])
y = energy_results["Total Jobs"]

#select the data and presentation
plt.scatter(x=x, y=y, s=40, alpha=1)
    #s is for point size; alpha is for color contrast; c is for color

#set the axis labels
ax.set_xlabel("Average Savings")
ax.set_ylabel("Total Jobs")

#set title
ax.set_title("Jobs versus Savings \n log scale")
```

```python
##histogram savings log scale

#open the plot
fig, ax = plt.subplots(figsize=(4, 4))

#define the variable
savings_log = np.log(energy_results["Average Savings"])

#plot the histogram
n, bins, patches = plt.hist(savings_log, bins='auto', color='#0504aa', alpha=0.7, rwidth=0.85)

#set axis labels
ax.set_xlabel("Total savings")
ax.set_ylabel("Frequency")

#set title
ax.set_title("Histogram average savings \n log scale")
```

```python
##calculate total jobs across months

#look for the date column
print([col for col in energy_data.columns])

#list dates
list_dates = list(energy_data["Effective Date"])

#list of months
list_months = []
for date in list_dates: #for each date
    
    #split by "/"
    dates_splitted = date.split("/")
    
    #select the month and append
    list_months.append(dates_splitted[0])
    
    #run this instead of the preivous line in case you want to calculate jobs per each month and year, i.e., jan-2020, jan-2021....
    #list_months.append(("/".join([dates_splitted[0], dates_splitted[-1]])))

#add as a new column
energy_data["Effective Month"] = list_months

#unique months
list_months = list(set(energy_data["Effective Month"]))

#order by month
list_months.sort(key=lambda date: datetime.strptime(date, "%m"))
    #convert date strings to datetime object and use it to sort

#create df to save results
results = pd.DataFrame(list_months)
results = results.rename(columns={0: "Effective Month"})
results["Total Jobs"] = ""

#calculate the total number of jobs created in each month
for month in list_months:
    
    #extract the job created in date
    jobs_selected_date = energy_data.loc[energy_data["Effective Month"] == month, "Job created"]
    
    #sum
    total_jobs = np.nansum(jobs_selected_date)
        #use np.nan just in case there are nan
    
    #save
    results.loc[results["Effective Month"] == month, "Total Jobs"] = total_jobs


#see results
results
```

```python
##plot the results

#open the plot
fig, ax = plt.subplots(figsize=(4, 4))

#select the data and presentation
plt.plot("Effective Month", "Total Jobs", data=results)
    #s is for point size; alpha is for color contrast; c is for color

#set the axis labels
ax.set_xlabel("Month")
ax.set_ylabel("Total Jobs")

#set title
ax.set_title("Jobs across Months")
```
